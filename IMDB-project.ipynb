### 20. Classify movies based on ratings

In this step, we classify movies into categories (Excellent, Good, and Average) based on their ratings, allowing us to group films according to their perceived quality.


```python
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

### 22. Find unique values from genre

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1[:20]
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```
